```python
import itertools
import random
import numpy as np
import torch
import pandas as pd

seed= 15
random.seed(seed)
np.random.seed(seed)

from transformers import set_seed

set_seed(seed)
```

```python
from datasets import load_dataset, load_metric

metric = load_metric('glue', 'cola')
```

#train model

```python
from datasets import load_from_disk
```

```python
ngrams_train = load_from_disk('data/cola/cola_train_ngrams')
```

```python
ngrams_dev= load_from_disk('data/cola/cola_val_ngrams')
```

```python
model_checkpoint = "bert-base-uncased"
batch_size = 16
```

```python
task='cola'
```

```python
from transformers import AutoTokenizer
    
tokenizer = AutoTokenizer.from_pretrained(model_checkpoint, use_fast=True)
```

```python
task_to_keys = {
    "cola": ("sentence", None),
    "mnli": ("premise", "hypothesis"),
    "mnli-mm": ("premise", "hypothesis"),
    "mrpc": ("sentence1", "sentence2"),
    "qnli": ("question", "sentence"),
    "qqp": ("question1", "question2"),
    "rte": ("sentence1", "sentence2"),
    "sst2": ("sentence", None),
    "stsb": ("sentence1", "sentence2"),
    "wnli": ("sentence1", "sentence2"),
}
```

```python
sentence1_key, sentence2_key = task_to_keys[task]
def preprocess_function(examples):
    if sentence2_key is None:
        return tokenizer(examples[sentence1_key], truncation=True)
    return tokenizer(examples[sentence1_key], examples[sentence2_key], truncation=True)
```

```python
encoded_train = ngrams_train.map(preprocess_function, batched=True)
```

```python
encoded_val = ngrams_dev.map(preprocess_function, batched=True)
```

```python
from transformers import AutoModelForSequenceClassification, TrainingArguments, Trainer

num_labels = 3 
model = AutoModelForSequenceClassification.from_pretrained(model_checkpoint, num_labels=num_labels)
```

```python
metric_name = "pearson" if task == "stsb" else "matthews_correlation" if task == "cola" else "accuracy"
model_name = model_checkpoint.split("/")[-1]

args = TrainingArguments(
    evaluation_strategy = "epoch",
    save_strategy = "epoch",
    learning_rate=2e-5,
    per_device_train_batch_size=batch_size,
    per_device_eval_batch_size=batch_size,
    num_train_epochs=5,
    weight_decay=0.01,
    load_best_model_at_end=True,
    metric_for_best_model=metric_name,
    output_dir=f'bert_base_finetuned_cola_ngrams_{seed}'
#    push_to_hub=True,
)
```

```python
acc= load_metric('accuracy')
```

```python
def compute_metrics(eval_pred):
    predictions, labels = eval_pred
    if task != "stsb":
        predictions = np.argmax(predictions, axis=1)
    else:
        predictions = predictions[:, 0]
    return  metric.compute(predictions=predictions, references=labels)
```

```python
def compute_acc(eval_pred):
    predictions, labels = eval_pred
    
    predictions = np.argmax(predictions, axis=1)

    return  acc.compute(predictions=predictions, references=labels)
```

```python
trainer = Trainer(
    model,
    args,
    train_dataset=encoded_train,
    eval_dataset=encoded_val,
    tokenizer=tokenizer,
    compute_metrics=compute_metrics
)
```

```python
trainer.train()
```

#evaluation

```python
datasets=[dict(path='data/cola/cola_dev_1gram' , name='cola_1g'),
          dict(path='data/cola/cola_dev_2gram' , name='cola_2g'),
          dict(path='data/cola/cola_dev_3gram' , name='cola_3g'),   
]

runs=[dict(path='bert_base_finetuned_cola_ngrams_15/checkpoint-4520/',
           name='invalid', seed=15),
      ]
```

```python
def load_model(path):
  tokenizer = AutoTokenizer.from_pretrained(path)
  model = AutoModelForSequenceClassification.from_pretrained(path)
  model_checkpoint=path
  model_name = model_checkpoint.split("/")[-2]
  args = TrainingArguments(
    f"{model_name}",
    evaluation_strategy = "epoch",
    save_strategy = "epoch",
    do_train= False,
    do_eval=  True,
    per_device_eval_batch_size=16,
  )
  return tokenizer, model, args


def eval_dataset(dataset, tokenizer, model, args):
  encoded_dataset= dataset.map(preprocess_function, batched=True)
  trainer = Trainer(
    model,
    args,
    train_dataset=None,
    eval_dataset=encoded_dataset,
    tokenizer=tokenizer,
    compute_metrics=compute_acc,
  ) 
  metrics= trainer.evaluate()
  return metrics

def eval_dataset_cola(tokenizer, model, args):
  dataset= load_dataset('glue', 'cola')['validation']
  encoded_dataset= dataset.map(preprocess_function, batched=True)
  trainer = Trainer(
    model,
    args,
    train_dataset=None,
    eval_dataset=encoded_dataset,
    tokenizer=tokenizer,
    compute_metrics=compute_acc,
  ) 
  metrics= trainer.evaluate()
  return metrics
```

```python
results_all={}
for run in runs:
  tokenizer, model, args= load_model(run['path'])
  results_all[f'seed_{run["seed"]}']={}
  temp_acc= eval_dataset_cola(tokenizer, model, args)['eval_accuracy']
  results_all[f'seed_{run["seed"]}']['dev']=temp_acc
  for dp in datasets:
    dataset= load_from_disk(dp['path'])
    temp_acc= eval_dataset(dataset, tokenizer, model, args)['eval_accuracy']
    results_all[f'seed_{run["seed"]}'][dp['name']]=temp_acc


```

```python
df=pd.DataFrame(results_all)
```
