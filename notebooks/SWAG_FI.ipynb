```python
import itertools
import random
import numpy as np
import torch
import pandas as pd

seed= 42
random.seed(seed)
np.random.seed(seed)

from transformers import set_seed

set_seed(seed)
```

#train model

```python
model_checkpoint = "bert-base-uncased"
batch_size = 16
```

```python
from datasets import load_dataset, load_metric, load_from_disk
```

```python
ngrams_train= load_from_disk('data/swag/swag_train_ngrams')
```

```python
ngrams_val= load_from_disk('data/swag/swag_val_ngrams')
```

```python
from transformers import AutoTokenizer
    
tokenizer = AutoTokenizer.from_pretrained(model_checkpoint, use_fast=True)
```

```python
#5 answer choices for the invalid training
ending_names = ["ending0", "ending1", "ending2", "ending3", "ending4"]

def preprocess_function(examples):
    # Repeat each first sentence four times to go with the four possibilities of second sentences.
    first_sentences = [[context] * 5 for context in examples["sent1"]]
    # Grab all second sentences possible for each context.
    question_headers = examples["sent2"]
    second_sentences = [[f"{header} {examples[end][i]}" for end in ending_names] for i, header in enumerate(question_headers)]
    
    # Flatten everything
    first_sentences = sum(first_sentences, [])
    second_sentences = sum(second_sentences, [])
    
    # Tokenize
    tokenized_examples = tokenizer(first_sentences, second_sentences, truncation=True)
    # Un-flatten
    return {k: [v[i:i+5] for i in range(0, len(v), 5)] for k, v in tokenized_examples.items()}
```

```python
encoded_val = ngrams_val.map(preprocess_function, batched=True)
```

```python
encoded_train = ngrams_train.map(preprocess_function, batched=True)
```

```python
from transformers import AutoModelForMultipleChoice, TrainingArguments, Trainer

model = AutoModelForMultipleChoice.from_pretrained(model_checkpoint)
```

```python
model_name = model_checkpoint.split("/")[-1]
args = TrainingArguments(
    evaluation_strategy = "epoch",
    save_strategy = "epoch",
    learning_rate=5e-5,
    per_device_train_batch_size=batch_size,
    per_device_eval_batch_size=batch_size,
    num_train_epochs=3,
    weight_decay=0.01,
    output_dir=f'bert_base_finetuned_swag_ngrams_seed_{seed}'
)
```

```python
from dataclasses import dataclass
from transformers.tokenization_utils_base import PreTrainedTokenizerBase, PaddingStrategy
from typing import Optional, Union
import torch

@dataclass
class DataCollatorForMultipleChoice:
    """
    Data collator that will dynamically pad the inputs for multiple choice received.
    """

    tokenizer: PreTrainedTokenizerBase
    padding: Union[bool, str, PaddingStrategy] = True
    max_length: Optional[int] = None
    pad_to_multiple_of: Optional[int] = None

    def __call__(self, features):
        label_name = "label" if "label" in features[0].keys() else "labels"
        labels = [feature.pop(label_name) for feature in features]
        batch_size = len(features)
        num_choices = len(features[0]["input_ids"])
        flattened_features = [[{k: v[i] for k, v in feature.items()} for i in range(num_choices)] for feature in features]
        flattened_features = sum(flattened_features, [])
        
        batch = self.tokenizer.pad(
            flattened_features,
            padding=self.padding,
            max_length=self.max_length,
            pad_to_multiple_of=self.pad_to_multiple_of,
            return_tensors="pt",
        )
        
        # Un-flatten
        batch = {k: v.view(batch_size, num_choices, -1) for k, v in batch.items()}
        # Add back labels
        batch["labels"] = torch.tensor(labels, dtype=torch.int64)
        return batch
```

```python
import numpy as np

def compute_metrics(eval_predictions):
    predictions, label_ids = eval_predictions
    preds = np.argmax(predictions, axis=1)
    return {"accuracy": (preds == label_ids).astype(np.float32).mean().item()}
```

```python
trainer = Trainer(
    model,
    args,
    train_dataset=encoded_train,
    eval_dataset=encoded_val,
    tokenizer=tokenizer,
    data_collator=DataCollatorForMultipleChoice(tokenizer),
    compute_metrics=compute_metrics,
)
```

```python
trainer.train()
```

#evaluation

```python
datasets=[dict(path='data/swag/swag_dev_sh_p_3gram_4' , name='3gram s1'),
          dict(path='data/swag/swag_dev_sh_p_2gram_4' , name='2gram s1'),
          dict(path='data/swag/swag_dev_sh_p_1gram_4' , name='1gram s1'),
          dict(path='data/swag/swag_dev_sh_q_3gram_4' , name='3gram endings'),
          dict(path='data/swag/swag_dev_sh_q_2gram_4' , name='2gram endings'),
          dict(path='data/swag/swag_dev_sh_q_1gram_4' , name='1gram endings'),
         
]
runs=[dict(path='bert_base_finetuned_swag_ngrams_seed_42/checkpoint-23829/' , name='invalid', seed=42)
]
```

```python
#eval model
from datasets import load_from_disk, load_dataset, load_metric
from transformers import AutoTokenizer, AutoModelForMultipleChoice, TrainingArguments, Trainer

ending_names = ["ending0", "ending1", "ending2", "ending3"]

def preprocess_function(examples):
    # Repeat each first sentence four times to go with the four possibilities of second sentences.
    first_sentences = [[context] * 4 for context in examples["sent1"]]
    # Grab all second sentences possible for each context.
    question_headers = examples["sent2"]
    second_sentences = [[f"{header} {examples[end][i]}" for end in ending_names] for i, header in enumerate(question_headers)]
    
    # Flatten everything
    first_sentences = sum(first_sentences, [])
    second_sentences = sum(second_sentences, [])
    
    # Tokenize
    tokenized_examples = tokenizer(first_sentences, second_sentences, truncation=True)
    # Un-flatten
    return {k: [v[i:i+4] for i in range(0, len(v), 4)] for k, v in tokenized_examples.items()}


def compute_metrics(eval_predictions):
    predictions, label_ids = eval_predictions
    preds = np.argmax(predictions, axis=1)
    return {"accuracy": (preds == label_ids).astype(np.float32).mean().item()}
```

```python
from dataclasses import dataclass
from transformers.tokenization_utils_base import PreTrainedTokenizerBase, PaddingStrategy
from typing import Optional, Union
import torch

@dataclass
class DataCollatorForMultipleChoice:
    """
    Data collator that will dynamically pad the inputs for multiple choice received.
    """

    tokenizer: PreTrainedTokenizerBase
    padding: Union[bool, str, PaddingStrategy] = True
    max_length: Optional[int] = None
    pad_to_multiple_of: Optional[int] = None

    def __call__(self, features):
        label_name = "label" if "label" in features[0].keys() else "labels"
        labels = [feature.pop(label_name) for feature in features]
        batch_size = len(features)
        num_choices = len(features[0]["input_ids"])
        flattened_features = [[{k: v[i] for k, v in feature.items()} for i in range(num_choices)] for feature in features]
        flattened_features = sum(flattened_features, [])
        
        batch = self.tokenizer.pad(
            flattened_features,
            padding=self.padding,
            max_length=self.max_length,
            pad_to_multiple_of=self.pad_to_multiple_of,
            return_tensors="pt",
        )
        
        # Un-flatten
        batch = {k: v.view(batch_size, num_choices, -1) for k, v in batch.items()}
        # Add back labels
        batch["labels"] = torch.tensor(labels, dtype=torch.int64)
        return batch
```

```python
def load_model(path):
  tokenizer = AutoTokenizer.from_pretrained(path, use_fast=True)
  model = AutoModelForMultipleChoice.from_pretrained(path)
  model_checkpoint=path
  model_name = model_checkpoint.split("/")[-2]
  args = TrainingArguments(
    f"{model_name}",
    evaluation_strategy = "epoch",
    save_strategy = "epoch",
    do_train= False,
    do_eval=  True,
    per_device_eval_batch_size=16,
  )
  return tokenizer, model, args


def eval_dataset(dataset, tokenizer, model, args):
  encoded_dataset= dataset.map(preprocess_function, batched=True)
  trainer = Trainer(
    model,
    args,
    train_dataset=None,
    eval_dataset=encoded_dataset,
    tokenizer=tokenizer,
    data_collator=DataCollatorForMultipleChoice(tokenizer),
    compute_metrics=compute_metrics,
  ) 
  metrics= trainer.evaluate()
  return metrics


def eval_swag(tokenizer, model, args):
  dataset= load_dataset('swag')['validation']
  encoded_dataset= dataset.map(preprocess_function, batched=True)
  trainer = Trainer(
    model,
    args,
    train_dataset=None,
    eval_dataset=encoded_dataset,
    tokenizer=tokenizer,
    data_collator=DataCollatorForMultipleChoice(tokenizer),
    compute_metrics=compute_metrics,
  ) 
  metrics= trainer.evaluate()
  return metrics
```

```python
results_all={}
for run in runs:
  tokenizer, model, args= load_model(run['path'])
  results_all[f'seed_{run["seed"]}']={}
  temp_acc= eval_swag(tokenizer, model, args)['eval_accuracy']
  results_all[f'seed_{run["seed"]}']['dev']=temp_acc
  for dp in datasets:
    dataset= load_from_disk(dp['path'])
    temp_acc= eval_dataset(dataset, tokenizer, model, args)['eval_accuracy']
    results_all[f'seed_{run["seed"]}'][dp['name']]=temp_acc


```
